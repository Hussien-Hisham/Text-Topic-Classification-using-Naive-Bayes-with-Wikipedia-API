**1. Loading Data: Download 5 different documents at least for training in 2 different domains (5 documents per each domain) and one document for test in each domain using Wikipedia API**

```python
def download_wikipedia_documents(domain, num_docs):
    wiki = wikipediaapi.Wikipedia('en')
    documents = []
    for i in range(num_docs):
        page = wiki.page(domain)
        document = page.text.split("\n")[1] # get first paragraph
        documents.append(document)
    return documents
```

```python
# Downloading 5 documents each for two different domains
domain1 = "Artificial Intelligence"
domain2 = "Football"
num_docs = 5

docs1 = download_wikipedia_documents(domain1, num_docs)
docs2 = download_wikipedia_documents(domain2, num_docs)

# Downloading one test document for each domain
test_doc1 = download_wikipedia_documents(domain1,1)
test_doc2 = download_wikipedia_documents(domain2,1)
```

**2. class_prob (doc_label)**

```python
def class_prob(labels):
    total_docs = len(labels)
    class_probabilities = {}
    for label in set(labels):
        class_docs = labels.count(label)
        class_probabilities[label] = class_docs / total_docs
    return class_probabilities
```

**3. Function preprocessing (doc)**

```python
def preprocessing(document):
    # Convert to lowercase
    document = document.lower()
    
    # Remove punctuations
    document = re.sub(r'[^\w\s]','',document)
    
    # Tokenize document
    tokens = document.split()
    
    # Remove stopwords
    stopwords = ["the", "and", "a", "an", "in", "to", "for", "is", "of", "on", "that", "with", "by", "at", "this", "it", "from", "or", "be", "as", "are", "was", "which", "can", "if", "there", "but", "not", "has", "had", "have", "will", "would", "should", "could", "may", "might", "been", "were", "who", "whom", "where", "what", "when", "why", "how", "about", "out", "into", "up", "down", "some", "more", "most", "such", "than", "then", "same", "other", "only", "over", "under", "its", "their", "our", "yours", "mine", "your", "his", "her", "hers", "them", "they", "us", "we", "i", "me", "my"]
    tokens = [token for token in tokens if token not in stopwords]
    
    return tokens
```

```python
preprocessed_docs1 = []
preprocessed_docs2 = []

for doc in docs1:
    preprocessed_docs1.append(preprocessing(doc))
    
for doc in docs2:
    preprocessed_docs2.append(preprocessing(doc))
```

**5. Function conditional_prob ()**

```python
def conditional_prob(word, doc_class, docs, all_tokens):
    class_docs = [doc for doc, label in docs if label == doc_class]
    num_docs_in_class = len(class_docs)
    num_docs = len(docs)
    num_tokens_in_class = sum([len(doc) for doc in class_docs])
    num_unique_tokens_in_class = len(all_tokens.intersection(set(word)))
    
    # Apply Laplace smoothing
    probability = (num_unique_tokens_in_class + 1) / (num_tokens_in_class + len(all_tokens) + 1)
    
    return probability
```

**6. Use the above methods to train your model and save to hard disk the probabilities
of all classes and Dictionary with key and value->{(class, word): prob}**

```python
def train_model(docs, labels):
    # Get probability of each class (domain)
    domain_probabilities = class_prob(labels)
    
    # Get preprocessed docs
    preprocessed_docs = []
    for doc in docs:
        preprocessed_docs.append(preprocessing(doc))
        
    # Get unique words in all documents
    all_tokens = set()
    for doc in preprocessed_docs:
        all_tokens.update(getTokens(doc))
    
    # Calculate conditional probabilities for each word in each class
    word_probs = {}
    for doc, label in zip(preprocessed_docs, labels):
        for word in doc:
            word_probs[(label, word)] = conditional_prob(word, label, list(zip(preprocessed_docs, labels)), all_tokens)
    
    # Save probabilities to hard disk
    with open('class_prob.pickle', 'wb') as f:
        pickle.dump(domain_probabilities, f)
        
    with open('word_prob.pickle', 'wb') as f:
        pickle.dump(word_probs, f)
```

```python
train_model(docs1 + docs2, ["Football"] * len(docs1) + ["Artificial Intelligence"] * len(docs2))
```

**7. Function predict(test_doc)**

```python
def predict(test_doc):
    # Load probabilities from hard disk
    with open('class_prob.pickle', 'rb') as f:
        domain_probabilities = pickle.load(f)
        
    with open('word_prob.pickle', 'rb') as f:
        word_probs = pickle.load(f)
    
    # Preprocess test document
    preprocessed_test_doc = preprocessing(test_doc)
    
    # Calculate probability of each class for test document
    class_probabilities = {}
    for domain, domain_prob in domain_probabilities.items():
        class_probabilities[domain] = domain_prob
        for word in preprocessed_test_doc:
            if (domain, word) in word_probs:
                class_probabilities[domain] *= word_probs[(domain, word)]
    
    # Return predicted class
    return max(class_probabilities, key=class_probabilities.get)
```
